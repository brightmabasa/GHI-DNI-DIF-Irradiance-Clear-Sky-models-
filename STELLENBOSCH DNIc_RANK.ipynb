```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the data from the CSV file
df = pd.read_csv('STEL_DNIc_metrics_results.csv')

# Handle NaN, infinity, or large values in the dataset
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)


#import io
# Read the data into a pandas DataFrame
#df = pd.read_csv(io.StringIO(data), sep='\t')
```

```python
# Calculate the absolute value of MBE
df['|MBE|'] = df['MBE'].abs()

# Rank the data based on |MBE|, MAE, and RMSE
df['Rank_MBE'] = df['|MBE|'].rank()
df['Rank_MAE'] = df['MAE'].rank()
df['Rank_RMSE'] = df['RMSE'].rank()

# Calculate the mean rank for each row
df['Mean_Rank'] = df[['Rank_MBE', 'Rank_MAE', 'Rank_RMSE']].mean(axis=1)

# Sort the DataFrame by Mean_Rank
df_sorted = df.sort_values(by='Mean_Rank')

# Select relevant columns for display and include model name
df_sorted.insert(0, 'Model', df_sorted.index)

# Round off the results to 2 decimal digits
df_sorted = df_sorted.round(2)

# Add a numeric rank column based on Mean_Rank starting from 1
df_sorted['Numeric_Rank'] = df_sorted['Mean_Rank'].rank(method='min').astype(int)
df['Numeric_Rank']= df_sorted['Numeric_Rank']

# Save the results to a CSV file
#df_sorted.to_csv('ranking_results.csv', index=False)

# Display the DataFrame with model names and rankings
#print(df_sorted[['Model', '|MBE|', 'MAE', 'RMSE', 'Rank_MBE', 'Rank_MAE', 'Rank_RMSE', 'Mean_Rank', 'Numeric_Rank']])
df_sorted
```
